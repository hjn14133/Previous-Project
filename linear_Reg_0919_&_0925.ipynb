```
>>> import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.utils import check_arrays
>>> 
# sum of sensor 1 and sensor 2
# raw sum of sensor 1
s1 = np.array([896.781, 555.786, 625.24,694.249,440.214, 802.261 ]).reshape(-1, 1)
# raw sum of sensor 2
s2 = np.array([9353.48, 8354.046, 10847.337, 14279.194,11771.49, 22746.853]).reshape(-1, 1)

# z_score of sensor 1 sum
s1_z = np.array([41.035, 35.709, 48.852,27.527, 12.657, 5.099]).reshape(-1, 1)
# z_score of sensor 2 sum
s2_z = np.array([602.123, 633.074, 626.557, 131.544, 49.479, 19.566]).reshape(-1, 1)

# min of sensor 1 sum (100)
s1_m = np.array([146.906, 112.2, 109.773, 82.479, 46.564, 39.452]).reshape(-1, 1)
# min of sensor 2 sum (100)
s2_m = np.array([1654.63, 1665.081, 1613.59, 341.486, 143.078, 77.233]).reshape(-1, 1)

# Actual weight of seed cotton (lbs)
AW = np.array([40, 40, 40, 35, 10, 5]).reshape(-1, 1)


reg1 = LinearRegression().fit(s1_z, AW)
reg2 = LinearRegression().fit(s2_z, AW)


AW1_pred = reg1.predict(s1_z)
AW2_pred = reg2.predict(s2_z)
# The coefficients
print('Coefficients of sensor 1: \n', reg1.coef_)
# The mean squared error
print("Mean squared error of sensor 1: %.2f"
      % mean_squared_error(AW, AW1_pred))
# Explained variance score: 1 is perfect prediction
print('R2 score of sensor 1: %.2f' % r2_score(AW, AW1_pred))
# The mean abosult error
print("Mean absolute error of sensor 1: %.2f"
      % mean_absolute_error(AW, AW1_pred))

# Plot outputs
plt.scatter(s1_z, AW,  color='black')
plt.plot(s1_z, AW1_pred, color='blue', linewidth=3)

plt.xlabel('Sensor 1 SUM with zscore algo')
plt.ylabel('Weight Data (lbs)')
plt.title('Lab Test in ginning lab')
#plt.xticks(())
#plt.yticks(())

plt.show()
```
